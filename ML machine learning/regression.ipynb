```python
#regression test for
# example of a regression test for a specific issue in the codebase
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regression Test')
plt.show()
```

```python
#model fitting and prediction
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
predictions = model.predict(x.reshape(-1, 1))
print(predictions)

```

```python
# y = m * x + c
m = model.coef_[0]
c = model.intercept_
print(f"Model parameters: m = {m}, c = {c}")
```

```python
#drawing the regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, predictions, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regression Test with Line')
plt.legend()
plt.show()
```

```python
year_of_experience = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
salary = np.array([30000, 40000, 50000, 60000, 70000])
model.fit(year_of_experience, salary)
predicted_salary = model.predict(year_of_experience)
print(predicted_salary)
plt.scatter(year_of_experience, salary, color='blue', label='Data Points')
plt.plot(year_of_experience, predicted_salary, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.legend()
plt.show()


```

```python
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor()
tree_model.fit(year_of_experience, salary)
tree_predictions = tree_model.predict(year_of_experience)
print(tree_predictions)
plt.scatter(year_of_experience, salary, color='blue', label='Data Points')
plt.plot(year_of_experience, tree_predictions, color='green', label='Decision Tree Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience with Decision Tree')
plt.legend()
plt.show()
```
