```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='*', color='blue')
plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = ["green", "blue", "orange", "red", "purple"]
y = [5, 7, 3, 8, 6]
plt.bar(x, y, color='orange')
plt.title('Sample Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [20, 30, 15, 25, 10]
plt.pie(y, labels=x, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Sample Pie Chart')
plt.show()
```

```python
import numpy as np
x = np.random.rand(100,1)
plt.hist(x, bins=10, color='cyan', edgecolor='black')
plt.title('Sample Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
x = np.random.rand(45,45)
plt.imshow(x, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Sample Heatmap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
