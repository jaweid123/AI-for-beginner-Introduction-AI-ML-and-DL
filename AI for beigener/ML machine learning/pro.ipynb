create class define

```python
class car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    def display_info(self):
        print(f"Car: {self.make} {self.model} ({self.year})")
    def behavior(self):
        print(f"The car {self.model} is driving.")    

C = car("Toyota", "lancrozar", 2020)
C.display_info()
C.behavior()
```
